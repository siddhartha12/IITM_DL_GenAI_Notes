# 4.2 - Transfer Learning

Problem: Training a large CNN from scratch requires massive datasets and huge computational resources

## Transfer Learning
process of taking a model pre-trained on a large dataset and fine tuning for a new specific task
* Use the pre trained convolutional base as a fixes feature extractor
* Remove the original classifier head
* add a new smaller classifier head for yout task
* Train only the new classifier head on your smaller custom dataset
This is the standard and most effective approach for most practical computer vision problem

## Various fomrs of Transfer Learning
1. As a feature Extractor
    * The pre trained convolutional base is used as a fixed feature extractor
    * We Freeze the weights of the base layers remove the original calssifier and add a new classifier
    * Only the new classifier is trainer, idea for new dataset small
2. as a fine-tuner
    * Initialize the network with pre-trained weights just like before
    * WE unfreeze either the entire network or the top few convolutional layers
    * We continue training the whole model (or part) end to end on the new data using a very low learning rate

**Why it works**: pre trained model already learned a powerful heirarchy of visual features, from universal edges and textures in early layers to complex object parts in deeper layers


## Transfer Learning for Segmentation
for segmentation, transfer learning is used to build the encoder part of an encoder-decoder network
(think of encoder-decoder like PCA, we encode values into a smaller set, then we can decode into the orginal values)

* **Using the pre trained model** -> entire encoder path is replaced with a pre-trained classification network, This is called the "backbone", this provides the rich semantic feature extraction learned from ImageNet
* **Training the decoder** -> new decoder head is added, takes the powerful low resolution features from the encoder and progressively upsamples them to rebuild the high resolution pixel level segmentation mask
